```python
df['Month'] = pd.to_datetime(df['Month'], format='%m-%d').apply(lambda x: x.replace(year=2023))
```

```python
df
```

```python
df['year'] = df['Month'].dt.year
df['month'] = df['Month'].dt.month
df['day'] = df['Month'].dt.day
df['dayofweek'] = df['Month'].dt.dayofweek  # Monday=0, Sunday=6
df['weekofyear'] = df['Month'].dt.isocalendar().week
df['quarter'] = df['Month'].dt.quarter
# ... add other relevant features like is_weekend, is_holiday, etc.
```

```python
df
```

```python
df['lag_1'] = df['Sales'].shift(1)
df['lag_7'] = df['Sales'].shift(7)  # For weekly lags
df['lag_30'] = df['Sales'].shift(30)  # For monthly lags
# ... create more lags as needed
```

```python
df
```

```python
df['rolling_mean_7'] = df['Sales'].rolling(7).mean()
df['rolling_std_7'] = df['Sales'].rolling(7).std()
# ... add other rolling statistics like min, max, etc.
```

```python
df['expanding_mean'] = df['Sales'].expanding().mean()
df['expanding_max'] = df['Sales'].expanding().max()
```
