```python
x_axis = np.linspace(0,12,num=200)
seasonality = 1.2*np.sin(2*np.pi*x_axis/4)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_axis,seasonality,label="Seasonal Component",color="b")
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Example of Seasonal Signal")
plt.grid(True)
plt.legend()
plt.show()
```
